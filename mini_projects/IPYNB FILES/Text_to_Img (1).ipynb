```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
height, width = 200, 600
blank_image = np.ones((height, width, 3), np.uint8) * 255
```

```python
text = "MAVERICK"
font = cv2.FONT_HERSHEY_SIMPLEX
font_scale = 2
font_color = (0, 0, 0)
font_thickness = 3
```

```python
text_size = cv2.getTextSize(text, font, font_scale, font_thickness)[0]
text_x = (blank_image.shape[1] - text_size[0]) // 2
text_y = (blank_image.shape[0] + text_size[1]) // 2
```

```python
cv2.putText(blank_image, text, (text_x, text_y), font, font_scale, font_color, font_thickness)
```

```python
plt.imshow(cv2.cvtColor(blank_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
